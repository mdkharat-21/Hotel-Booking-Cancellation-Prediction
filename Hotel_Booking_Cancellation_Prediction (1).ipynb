```python
# Step-4) Feature Engineering:
#  In this step, the following preprocessing techniques are performed:
# 1. Missing value handling
# 2. Encoding categorical variables
# 3. Outlier handling
```

```python
# 1. Missing value handling
```

```python
df.isna().mean()*100
```

```python
# The 'company' feature has a high percentage of missing values (around 94%),
# so it is removed from the dataset.
```

```python
df.drop(['company'],axis=1,inplace=True)
```

```python
# The 'agent' column contains continuous numerical values,
# so missing values are filled using the median.
df['agent'].fillna(df['agent'].median(),inplace=True)
```

```python
df['agent'].isna().sum()
```

```python
#country column has categorical values so it filled using mode()
df['country'].fillna(df['country'].mode()[0],inplace=True)
```

```python
df['country'].isna().sum()
```

```python
#this column has categorical values it filled using mode()
df['children'].mode()
df['children'].fillna(df['children'].mode()[0],inplace=True)
```

```python
df['children'].isna().sum()
```

```python
# Dropping columns causing data leakage

# 'reservation_status' directly reveals whether the booking
# was canceled or not, which can lead to data leakage.

# 'reservation_status_date' contains future information
# that would not be available at the time of prediction.

# To build a realistic and unbiased prediction model,
# these columns are removed before training.

df.drop(['reservation_status','reservation_status_date'],axis=1,inplace=True)
```

```python
# 2. Encoding categorical variables
```

```python
df['hotel'].unique()
```

```python
df['hotel'].replace({'Resort Hotel':0, 'City Hotel':1},inplace=True)
```

```python
df['arrival_date_month'].unique()
```

```python
df['arrival_date_month'].replace({'July':0, 'August':1, 'September':2, 'October':3, 'November':4, 'December':5,
       'January':6, 'February':7, 'March':8, 'April':9, 'May':10, 'June':11},inplace=True)
```

```python
df['meal'].unique()
```

```python
df['meal'].replace({'BB':0, 'FB':1, 'HB':2, 'SC':3, 'Undefined':4},inplace=True)
```

```python
df['market_segment'].unique()
```

```python
df['market_segment'].replace({'Direct':0, 'Corporate':1, 'Online TA':2, 'Offline TA/TO':3,
       'Complementary':4, 'Groups':5, 'Undefined':6, 'Aviation':7},inplace=True)
```

```python
df['distribution_channel'].unique()
```

```python
df['distribution_channel'].replace({'Direct':0, 'Corporate':1, 'TA/TO':2, 'Undefined':3, 'GDS':4},inplace=True)
```

```python
df['reserved_room_type'].unique()
```

```python
df['reserved_room_type'].replace({'C':0, 'A':1, 'D':2, 'E':3, 'G':4, 'F':5, 'H':6, 'L':7, 'P':8, 'B':9},inplace=True)
```

```python
df['assigned_room_type'].unique()
```

```python
df['assigned_room_type'].replace({'C':0, 'A':1, 'D':2, 'E':3, 'G':4, 'F':5, 'I':6, 'B':7, 'H':8, 'P':9, 'L':10, 'K':11},inplace=True)
```

```python
df['deposit_type'].unique()
```

```python
df['deposit_type'].replace({'No Deposit':0, 'Refundable':1, 'Non Refund':2},inplace=True)
```

```python
df['customer_type'].unique()
```

```python
df['customer_type'].replace({'Transient':0, 'Contract':1, 'Transient-Party':2, 'Group':3},inplace=True)
```

```python
df['country'].unique()
```

```python
country_target = df.groupby('country')['is_canceled'].mean()
df['country'] = df['country'].map(country_target)
```

```python
#Step-4: Feature Selection
```

```python
x=df.drop(['is_canceled'],axis=1)
y=df['is_canceled']
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif

selector = SelectKBest(score_func=mutual_info_classif, k=10)

X_new = selector.fit_transform(x, y)

selected_features = x.columns[selector.get_support()]

print(selected_features)
```

```python
plt.figure(figsize=(8,10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr[['is_canceled']].sort_values(by='is_canceled',ascending=False),annot=True,cmap='coolwarm')
plt.show()
```
